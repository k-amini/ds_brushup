### Normal Distribution



There appears to be something slightly wrong with the normal one, it's not yielding the same height as the Wikipedia example I am stealing it from.

```python
x_lower_limit = -5
x_upper_limit = 5
x = np.linspace(x_lower_limit, x_upper_limit, 1000)
y_lower_limit = 0
y_upper_limit = 1

# The parameters are  and ^2
parameter_sets = [{"mu":0, "sigma_sq":0.2, "color":"blue"},
                  {"mu":0, "sigma_sq":1.0, "color":"red"},
                  {"mu":0, "sigma_sq":5.0, "color":"orange"},
                  {"mu":-2, "sigma_sq":0.5, "color":"green"} ]

fig1, ax1 = plt.subplots(figsize=(30, 10))

# The Normal distribution is described here
# [ https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html ]
for parameter_set in parameter_sets:
    #print(f" : {parameter_set['mu']}\t2 : {parameter_set['sigma_sq']}")
    ax1.plot(x, stats.norm.pdf(x,
                               parameter_set["mu"],
                               parameter_set["sigma_sq"]),
                               label = f" = {parameter_set['mu']:0.1f}, ^2 = {parameter_set['sigma_sq']:0.1f}",
                               color = parameter_set["color"],
                               linewidth=5)
plt.ylim(y_lower_limit, y_upper_limit)
plt.title('Normal Distribution', size=48)
plt.rc('xtick', labelsize=36) 
plt.rc('ytick', labelsize=36) 

leg = ax1.legend(prop={'size': 36})

filename = path + "fig_normal.png"

fig1.savefig(filename)
```

### Beta Distribution

```python
x_lower_limit = 0
x_upper_limit = 1
x = np.linspace(x_lower_limit, x_upper_limit, 1000)
y_lower_limit = 0
y_upper_limit = 2.5

# Using  and  as shape parameters
parameter_sets = [{"alpha":0.5, "beta":0.5, "color":"red"},
                  {"alpha":5,   "beta":1,   "color":"blue"},
                  {"alpha":1,   "beta":3,   "color":"green"},
                  {"alpha":2,   "beta":2,   "color":"purple"},
                  {"alpha":2,   "beta":5,   "color":"orange"} ]

fig1, ax1 = plt.subplots(figsize=(30, 10))

# The Beta distribution is described here
# [ https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html ]
for parameter_set in parameter_sets:
    ax1.plot(x, stats.beta.pdf(x,
                               parameter_set["alpha"],
                               parameter_set["beta"]),
                               label = f" = {parameter_set['alpha']:0.1f},  = {parameter_set['beta']:0.1f}",
                               color = parameter_set["color"],
                               linewidth=5)

plt.ylim(y_lower_limit, y_upper_limit)
plt.title('Beta Distribution', size=48)
plt.rc('xtick', labelsize=36) 
plt.rc('ytick', labelsize=36) 

leg = ax1.legend(prop={'size': 36})

filename = path + "fig_08.png"
fig1.savefig(filename)
```

### Gamma Distribution


Using the parameters as shape and rate :

```python
x = np.linspace(0, 20, 1000)
y_lower_limit = 0
y_upper_limit = 0.5

# Using k is shape and  is scale
parameter_sets = [{"k":1.0, "theta":2.0, "color":"red"},
                  {"k":2.0, "theta":2.0, "color":"orange"},
                  {"k":3.0, "theta":2.0, "color":"yellow"},
                  {"k":5.0, "theta":1.0, "color":"green"},
                  {"k":9.0, "theta":0.5, "color":"black"},
                  {"k":7.5, "theta":1.0, "color":"blue"},
                  {"k":0.5, "theta":1.0, "color":"purple"}
                 ]

fig1, ax1 = plt.subplots(figsize=(30, 10))

# Gamma distribution is described here:
# [ https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html ]
for parameter_set in parameter_sets:
    ax1.plot(x, stats.gamma.pdf(x,
                                parameter_set["k"]) / parameter_set["theta"],
                                label = f"k = {parameter_set['k']:0.1f},  = {parameter_set['theta']:0.1f}",
                                color = parameter_set["color"],
                                linewidth=5)
plt.ylim(y_lower_limit, y_upper_limit)
plt.title('Gamma Distribution (Shape & Scale)', size=48)
plt.rc('xtick', labelsize=36) 
plt.rc('ytick', labelsize=36) 

leg = ax1.legend(prop={'size': 36})

filename = path + "fig_gamma.png"
fig1.savefig(filename)
```

### Exponential Distribution

```python

y_lower_limit = 0
y_upper_limit = 1.5
x_lower_limit = 0
x_upper_limit = 5
x = np.linspace(x_lower_limit, x_upper_limit, 1000)

# Using  as a rate parameter
parameter_sets = [{"lambda":0.5, "color":"orange"},
                  {"lambda":1.0,   "color":"purple"},
                  {"lambda":1.5,   "color":"cyan"} ]

fig1, ax1 = plt.subplots(figsize=(30, 10))

# Expnential distribution function is described here :
# [ https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon ]
for parameter_set in parameter_sets:
    ax1.plot(x, stats.expon.pdf(x) * parameter_set["lambda"],
                               label = f" = {parameter_set['lambda']:0.1f}",
                               color = parameter_set["color"],
                               linewidth=5)
    
plt.ylim(y_lower_limit, y_upper_limit)
plt.title('Exponential Distribution (Rate)', size=48)
plt.rc('xtick', labelsize=36) 
plt.rc('ytick', labelsize=36) 

leg = ax1.legend(prop={'size': 36})

filename = path + "fig_exponentiall.png"
fig1.savefig(filename)
```
